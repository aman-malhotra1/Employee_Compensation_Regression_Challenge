```python
# Import important Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',100)
%matplotlib inline
```

```python
# Load Training and Test data
train_data = pd.read_csv('Clean_data_train.csv')
test_data = pd.read_csv('Clean_data_test.csv')

print(train_data.shape)
print(test_data.shape)
```

```python
# Display First Five record of test data
train_data.head()
```

```python
categorical_column = ['OG','Dept','Union','JF','Job']
numerical_column = ['Salaries','Overtime','H/D']
target_column = 'Total_Compensation'
```

```python
plt.figure(figsize=(18,8))
sns.countplot(train_data[categorical_column[0]], order=train_data[categorical_column[0]].value_counts().index)
plt.xticks(rotation = 90)
plt.title("No.Of Employees Representing Organization Group")
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.boxplot(x =train_data[categorical_column[0]],y= train_data[target_column],order=train_data.groupby('OG')['Total_Compensation'].median().sort_values(ascending=False).index)
plt.xticks(rotation = 90)
plt.title("No. Of Employees representing Organizational group Vs Total Compensation")
plt.show()
```

```python
top_30_jobs = train_data['Job'].value_counts().head(30)
plt.figure(figsize=(18,8))
sns.barplot(top_30_jobs.index,top_30_jobs.values)
plt.xticks(rotation = 90)
plt.title("No.Of Employees Representing Jobs")
plt.xlabel('JOBS')
plt.show()
```

```python
jobs = top_30_jobs.index
top_30_jobs_ = train_data[train_data['Job'].isin(jobs)]
plt.figure(figsize=(18,8))
sns.boxplot(x =top_30_jobs_['Job'],y= train_data[target_column],order=top_30_jobs_.groupby('Job')['Total_Compensation'].median().sort_values(ascending=False).index)
plt.xticks(rotation = 90)
plt.title("No. Of Employees representing Jobs Vs Total Compensation")
plt.show()
```
